# Solution for problem 1.1


```python
# Function to print a list
# Input: A list
# Output: Display each element in seperate line

def print_list(list):
    print("Printing List:") 
    for item in list: #For loop to iterate though the each item in the list
        print(item)   #Print the each item of the list  
        
# Testing function print_list with list1
list1 = ['a', 1, ['b', 'c']] 
print_list(list1)

print()

# Testing function print_list with list2
list2 = [True,False]
print_list(list2)
```

# Solution for problem 1.2

```python
int_list = []                  #Creating a empty list
int_list = list(range(0,6))    #Generate a list of integer from 0 to 5
print(int_list)                #Print the list
reversed_list = int_list[::-1] #Reverse the integer list and assign it another variable
print(reversed_list)           #Print the reversed list
```

###### Explaination: In order to reverse a list I used the slicing functionality of python. [::1] slices the list, goes through each items, by steps of -1. This generates a copy of list where the items are duplicated and the original list remains unchanged.  

# Part 2. Dictionaries

```python
def capitals_of_provinces(provinces, capital): #Function to print the province and repective capitals
    print('Required Output: ')
    for key,item in provinces.items():         #Interate though the disctionary of provinces            
        if item in capital:                    #Check if the shorthand of province is in 'capital' dictionary
            print('The capital of ', key,' is', capital.get(item))
                                               #Print the required output
        
provinces_dict = {'Ontario':'ON', 'Alberta':'AB' , 'Nova Scotia' : 'NS'} #Initial dictionary of provinces
capital_dict = {'ON':'Toronto', 'AB':'Edmonton'}  #Initial dictionary of capital

capitals_of_provinces(provinces_dict, capital_dict)  #Call the function

other_provinces_dict = {'British Columbia':'BC','Manitoba':'MB'} 
other_capitals_dict = {'BC':'Victoria','MB':'Winnipeg'}

provinces_dict.update(other_provinces_dict)       #Update the province dictionary with new values
capital_dict.update(other_capitals_dict)          #Update the capital dictionary with new values

print()
capitals_of_provinces(provinces_dict,capital_dict) #Call the function again to check the required output
    
```

# Part 3. Greatest Common Divisor (GCD)

```python
import math                                       # Import math package

def inbuilt_gdc(num1, num2):                       
    return(math.gcd(num1, num2))                  # Return the GDC of two number using in-built function
    
def custom_gdc(int1, int2):
    if int2 == 0:
        return int1                              # Return the maximum number if the other number is zero
    else:
        return custom_gdc(int2, int1 % int2)     # Recursive call taking 2nd number as 1st parameter 
                                                 # and modulus of both numbers as 2nd parameter
                                                 # The recursive call is made till the 2nd number is equal to zero.

#Testing
print('Inbuilt function result: ', inbuilt_gdc(20, 10), ' || Custom function result: ', custom_gdc(20, 10))
print('Inbuilt function result: ', inbuilt_gdc(3434, 1434), ' || Custom function result: ', custom_gdc(3434, 1434))
print('Inbuilt function result: ', inbuilt_gdc(50,123456), ' || Custom function result: ', custom_gdc(50,123456))
```

# Option 2 - Birthday paradox


```python
import random

num_students = 45                                    #Number of sample for students in each class
num_simulations = 1000                               #Number of simulation to run the program
print('Number of students: ', num_students)          #Printing the number of students
print('Number of simulations: ', num_simulations)    #Printing the number of simulations

def has_duplicates(t):
    """Returns True if any element appears more than once in (t),
    False otherwise."""
    s = t[:]                                         #Assign the list of students birthday to another variable, so that original list is not manipulated 
    s.sort()                                         #Sort the birthday list in accending order
    for i in range(len(s)-1):                        #For each item in the birthday list
        if s[i] == s[i+1]:                           #check if two of the birthday dates are same
            return True                              #if so, return True
    return False                                     #else, return false  

def list_of_duplicates(t):
    """Returns the list of duplicates."""
    s = t[:]                                         #Assign the list of students birthday to another variable, so that original list is not manipulated 
    duplicate_list = []                              #Initialize the list of duplicate birthdays
    s.sort()                                         #Sort the birthday list in accending order
    for i in range(len(s)-1):                        #For each item in the birthday list
        if s[i] == s[i+1]:                           #check if two of the birthday dates are same
            duplicate_list.append(s[i])              #Add the date to the list of duplicates birthday
    return duplicate_list                            #return the list of duplicate birthdays


def random_bdays(n):
    """Returns a list of integers between 1 and 365, with length (n)."""
    t = []
    for i in range(n):
        bday = random.randint(1, 365)                #For the number of simulations create a list of random birthday dates
        t.append(bday)                               #Add the random date to the list  
    return t                                         #Return the list of all birthday dates


def count_matches(students, samples):
    """Generates (samples) samples of (students) students, and counts
    how many of them have at least one pair of students with the same bday."""
    count = 0
    print('List of duplicates: ')
    for i in range(samples):
        t = random_bdays(students)                  #Generate the list of random birthday dates by calling random_bday function 
        duplicates = list_of_duplicates(t)          #Get the list of duplicates the for each simulation by calling list_of_duplicates functions      
        if duplicates:
            print('     Duplicate in simulation ', i , ' are ', duplicates) #Print the duplicates, if present 
        if has_duplicates(t):                       #Count the number of duplicates 
            count += 1
    return count                                    #Return the number of duplicates

"""run the birthday simulation 1000 times and print the number of matches and probability"""
count = count_matches(num_students, num_simulations)#Get the number of duplicates by calling count_matches functions
req_probability = count/num_simulations             #Calculate the probabilty by dividing the number of duplicates by the number of simulations    

print()
print('There were', count ,'simulations with at least one match') #Print the number of duplicates
print('The probability that two of the students have the same birthday: ', req_probability) #Print the probability of two of the students have the same birthday
```
